```r
install.packages("BiocManager")

BiocManager::install("ensembldb")
BiocManager::install("AnnotationHub")
BiocManager::install("GenomicRanges")
```

```r
library(ensembldb)
library(AnnotationHub)
library(GenomicRanges)
```

```r
ah <- AnnotationHub()

edb <- ah[["AH119325"]]
gene_symbol <- "PF4"
tx_df <- transcripts(
  edb,
  filter = GeneNameFilter(gene_symbol),
  return.type = "DataFrame"
)
tx_df
```

```r
write.table(
  as.data.frame(tx_df),
  file = "transcripts.tsv",
  sep = "\t",
  row.names = FALSE,
  quote = FALSE
)
pr_df <- proteins(
  edb,
  filter = GeneNameFilter(gene_symbol),
  return.type = "DataFrame"
)

pr_df
```

```r
write.table(
  as.data.frame(pr_df),
  file = "proteins.tsv",
  sep = "\t",
  row.names = FALSE,
  quote = FALSE
)
dir.create("protein_maps", showWarnings = FALSE)

for (i in seq_len(nrow(pr_df))) {

  pid  <- pr_df$protein_id[i]

  pseq <- pr_df$protein_sequence[i]

  if (is.na(pid) || is.na(pseq) || nchar(pseq) == 0)
    next

  rng <- IRanges(
    start = 1,
    end = nchar(pseq),
    names = pid
  )

  map <- proteinToGenome(rng, edb)[[1]]

  write.table(
    as.data.frame(map),
    file = file.path(
      "protein_maps",
      paste0(pid, "_map.tsv")
    ),
    sep = "\t",
    row.names = FALSE,
    quote = FALSE
  )
}
getwd()
```
